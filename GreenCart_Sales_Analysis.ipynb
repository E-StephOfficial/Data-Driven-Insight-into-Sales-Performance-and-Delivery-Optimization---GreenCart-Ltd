```python
# Fix gender and loyalty tier formatting
if 'gender' in customer_df.columns:
    customer_df['gender'] = customer_df['gender'].str.strip().str.title()

if 'loyalty_tier' in customer_df.columns:
    customer_df['loyalty_tier'] = customer_df['loyalty_tier'].str.strip().str.title()  # not upper anymore

if 'payment_method' in sales_df.columns:
    sales_df['payment_method'] = sales_df['payment_method'].str.strip().str.title()

# Function to fill NaNs smartly
def fill_missing_values(df):
    for col in df.columns:
        if df[col].isna().sum() > 0:
            if df[col].dtype == 'object':
                mode_val = df[col].mode()[0] if not df[col].mode().empty else 'Unknown'
                df[col].fillna(mode_val, inplace=True)
            else:
                median_val = df[col].median()
                df[col].fillna(median_val, inplace=True)
    return df

# Apply to all three datasets
customer_df = fill_missing_values(customer_df)
sales_df = fill_missing_values(sales_df)
product_df = fill_missing_values(product_df)
```

## Step 4: Merging the datasets together
To analyze sales and customer behavior effectively, we merged:

- `sales_df` with `customer_df` on `customer_id`

This gave us a single, unified dataset combining sales, customer, and product details.

## Step 4: Feature Engineering

In this step, we created new features to enrich the dataset and enable deeper analysis:

1. **Revenue**:  
   Calculated as `quantity  unit_price  (1 - discount_applied)` to account for discounts.

2. **Order Week**:  
   Extracted ISO week number from `order_date` using `.dt.isocalendar().week`.

3. **Price Band**:  
   Categorized `unit_price` into:
   - Low (< 15)
   - Medium (1530)
   - High (> 30)

4. **Days to Order**:  
   Calculated the number of days between `launch_date` and `order_date`.

5. **Email Domain**:  
   Extracted the domain from customer emails (e.g., gmail.com) using regex.

6. **Is Late**:  
   Created a boolean column that flags orders as `True` if `delivery_status` is Delayed.

These engineered features prepare the dataset for richer business insights in subsequent steps.


```python
# 1. Ensure quantity is numeric
merged_df['quantity'] = pd.to_numeric(merged_df['quantity'], errors='coerce')

# 2. revenue = quantity  unit_price  (1 - discount_applied)
merged_df['revenue'] = merged_df['quantity'] * merged_df['unit_price'] * (1 - merged_df['discount_applied'].fillna(0))

# 3. order_week = ISO week from order_date
merged_df['order_week'] = merged_df['order_date'].dt.isocalendar().week

# 4. price_band = Categorise unit_price as Low (<15), Medium (1530), High (>30)
merged_df['price_band'] = pd.cut(
    merged_df['unit_price'],
    bins=[-float('inf'), 15, 30, float('inf')],
    labels=['Low', 'Medium', 'High']
)

# 5. days_to_order = Days between launch_date and order_date
merged_df['launch_date'] = pd.to_datetime(merged_df['launch_date'], errors='coerce')
merged_df['days_to_order'] = (merged_df['order_date'] - merged_df['launch_date']).dt.days

# 6. email_domain = Extract domain from email
merged_df['email_domain'] = merged_df['email'].str.extract(r'@(.+)$')

# 7. is_late = True if delivery_status is "Delayed"
merged_df['is_late'] = merged_df['delivery_status'].str.lower().eq('delayed')
```

```python
product_perf = merged_df.groupby('category').agg({
    'revenue': 'sum',
    'quantity': 'sum',
    'discount_applied': 'mean'
}).reset_index().rename(columns={'discount_applied': 'avg_discount'})
```

```python
merged_df['signup_month'] = pd.to_datetime(merged_df['signup_date']).dt.month
```

```python
customer_behavior = merged_df.groupby(['loyalty_tier', 'signup_month']).agg({
    'revenue': 'sum',
    'order_id': 'nunique',
    'quantity': 'mean',
    'discount_applied': 'mean'
}).reset_index().rename(columns={
    'order_id': 'num_orders',
    'quantity': 'avg_quantity',
    'discount_applied': 'avg_discount'
})

customer_behavior.head()
```

```python
# Calculate delivery performance based on available columns
delivery_performance = merged_df.groupby(['region', 'price_band']).agg({
    'is_late': 'sum',                  # Total number of late deliveries
    'order_id': 'nunique'              # Total unique orders
}).reset_index().rename(columns={
    'is_late': 'num_late_deliveries',
    'order_id': 'total_orders'
})

# Calculate % late deliveries
delivery_performance['late_delivery_rate (%)'] = round(
    (delivery_performance['num_late_deliveries'] / delivery_performance['total_orders']) * 100, 2
)

delivery_performance.head()
```

```python
# Fix string 'nan' entries in payment_method
merged_df['payment_method'] = merged_df['payment_method'].astype(str).str.strip().str.title()

# Remove actual missing values and fake ones
merged_df = merged_df[~merged_df['payment_method'].isin(['Nan', 'NaN', 'None'])]  # Remove string NaNs
merged_df = merged_df.dropna(subset=['payment_method'])  # Remove real NaNs

# Standardize loyalty_tier
merged_df['loyalty_tier'] = merged_df['loyalty_tier'].str.strip().str.lower()
merged_df['loyalty_tier'] = merged_df['loyalty_tier'].replace({
    'gold': 'Gold',
    'gld': 'Gold',
    'silver': 'Silver',
    'sllver': 'Silver',
    'bronze': 'Bronze',
    'brnze': 'Bronze'
})
```

```python
payment_pref = merged_df.groupby(['loyalty_tier', 'payment_method']).agg({
    'customer_id': 'nunique'
}).reset_index().rename(columns={
    'customer_id': 'num_customers'
})

payment_pref_pivot = payment_pref.pivot(index='loyalty_tier', columns='payment_method', values='num_customers').fillna(0).astype(int)
payment_pref_pivot
```
